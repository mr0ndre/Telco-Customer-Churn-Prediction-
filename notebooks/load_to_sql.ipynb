```python
import pandas as pd 
import sqlite3
import os 
import matplotlib.pyplot as plt
from pathlib import Path 

PROJECT_ROOT = Path.cwd().parent

# ----------- Creating Customer Churn table --------------

churn_path = str(PROJECT_ROOT/'data'/'raw'/'Telco_customer_churn.xlsx')
db_path = 'churn.db'

df_temp = pd.read_excel(churn_path)

conn = sqlite3.connect(db_path)

# Loading into SQL 
df_temp.to_sql(name='customer_churn_raw', con=conn, if_exists='replace', index=False)

query = """
SELECT *
FROM customer_churn_raw
"""
df = pd.read_sql(query, conn)
df.head()

conn.close()
```
